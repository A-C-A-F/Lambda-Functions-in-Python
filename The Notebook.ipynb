### Task 1: What is Lambda Expression ? 

- Let's define a python function:  get two numerical variables as an input, and calculate the multiplication of two values and return it.  

```python
def mul(a,b):
    r = a * b
    return r
mul(4,5)
```

- let's define a lambda expression with the same purpose



```python
mul_lambda = lambda a,b : a * b
```

- let's check the variable type 

```python
mul_lambda
```

- let's call the lambda 

```python
mul_lambda(8,5)
```

- example : define a lambda function: gets a value and return square of the value

```python
square = lambda value : value ** 2
square(3)
```

- example : define lambda function: gets three values and return a list containing those values

```python
to_list = lambda v1, v2, v3 : [v1,v2,v3]
to_list(3,4,5)
```

- example : define lambda function : gets a string and return the first character of the string

```python
first_char = lambda string : string[0]
first_char("Hello")
```

- Exercise : define a lambda function: gets a string and return the length of it 

```python
get_length = lambda string : len(string)
get_length("Return the length")
```

### Task 2: Lambda Functions and Lists

- let's define a normal function: get's a list, return a list of all values to the power of 2 

```python
numbers = [1,2,3,10,9,3,4,10,7]
```

```python
def list_square(list_):
    new_list = []
    for element in list_:
        new_list.append(element**2)
    return new_list

list_square(numbers)    
```

- let's use map function 

```python
# this function we'll get a function as an input
# and a list as the output and what it does, it
# basically goes through this list and apply this function
# to each of the elements that we have in the list.

def func(element) :
    return element **2
map(func,numbers)
# If I run this it will return a map object.
```

```python
# But I can easily convert this map object into a list
# and I run this again.

list(map(func,numbers))

# So as we can see, we got the same result with list_square function.
# Now what we can do here is that instead of using a normal
# function, we can define this function as a lambda.
# And then use lambda in the map function.
```

- map function and lambda 

```python
func_lambda = lambda element : element ** 2
list(map(func_lambda, numbers))
# If I run this as we can see we have the same result.
```

- let's make it even more compact

```python
list(map(lambda element : element **2, numbers))

# So using the lambda expression, we could have wrote
# three lines of code in only one line of code.
```

- example: I have a list of words, let's use map function to convert all the words into upper case.

```python
words = ["football","python","java","coffee","keyboard"]
```

```python
list(map(lambda word : word.upper(), words))
```

- Exercise : we have a list of sentences, use map function and lambda to convert all the sentences into lower case. 

```python
sentences = [
    "Today is Thursday and weather Is Sunny.",
    "Tomorrow at 14 PM, Meeting with Daniel.",
    "Python is a powerfull Tool For DataScience.",
    "Tommy Called Today, He told me He does not feel Good... I think It's COVID!"
]
```

```python
list(map(lambda sentence : sentence.lower(), sentences))
```

### Task 3 : Lambda Expression combined with if statement

- let's define a normal function: gets a value and return true of the value is even, return false if the value is odd

```python
def isEven(value):
    if value%2==0:
        return True
    else:
        return False
isEven(4)        
```

- let's define a lambda with same purpose

```python
isEven_lambda = lambda value : True if value%2==0 else False
isEven_lambda(3)

# So in this way we can actually create lambda expressions
# combined with if and else statement.
```

- consider a list of numbers, define a normal function : gets a list of numbers and create a new list that contains only the even numbers from our input list

```python
numbers = [1,2,3,10,9,3,4,10,7]
```

```python
def filter_even(list_):
    new_list = []
    for element in list_:
        if element%2==0:
            new_list.append(element)
    return new_list

filter_even(numbers)
```

- let's use filter function 

```python
# So a filter function gets a function and a list.
# So what it does, again like map function, it goes through each
# of the elements of this list and will apply this function
# to each of the elements in this list.
# So it's basically the same as map function.
# But the difference here is that it will only keep those items
# in the list when we passed the item into the function,
# the result is true.
# So it only keeps or filter out those items that will return
# a true when we apply this function to them.

def func(element):
    if element % 2 == 0:
        return True
    else:
        return False

list(filter(func, numbers))

# If I run this I get only the even numbers in my list.

# So this function returns true if the element from this list
# is even or return false if the element of this list is odd.
# So this filter function will only keep those elements that
# returns it true after being applied to this function.
```

```python
# instead of this, we can easily use a lambda
# expression like as we used uh for the map function.

list(filter(lambda e : True if e % 2 == 0 else False, numbers))
# I run this and we got the same result.
```

- example: I have a list of names, let's use filter function and lambda to filter out the names starting with 'A'

```python
names = ["John","Alex","Maiki","Alice","Layla","Hanna","Alexander","Anil","Angela","Lola","Thelma","Clara"]
```

```python
list(filter(lambda name : True if name[0] == "A" else False, names))

# I run this as you can see, I have filtered out only
# those names that starts with a capital A .
```

- example: use filter and lambda function to filter out the names that starts with an 'A' and with the length of 4 

```python
list(filter(lambda name : True if (name[0] == "A") and (len(name) == 4) else False, names))
# so this lambda will only returns true if both
# of these conditions satisfied.

# So if I run this, I can see that I have only those names
# starting with A with the length of four.
```

- exercise: we have a list of numbers, use map function and lambda to multiply the odd values by 10 and the even values by 100 in our numbers list

```python
numbers = [1,2,3,10,9,3,4,10,7]
```

```python
list(map(lambda n : (n%2==0) and (n%2==1), numbers))
```

```python
list(map(lambda element : element *10 if element%2==1 else element *100, numbers))
```

### Task 4: Lambda expressions and sorting 

#### Sorting a list of tuples 
- Let's say we have a list of products with their respective prices, stored as a tupple in a list.

```python
products = [
    ("Cheese",1.50),
    ("Banana Bread",2.10),
    ("Chicken Meat",3.50),
    ("Apple Juice",0.45),
    ("Orange Juice",1.25),
    ("Milk",0.75),
    ("Mango",1.25)
]
```

- Let's sort the list based on the prices 

```python
# if I run the code below, we can see that all of the products are
# sorted based upon the name of the product, so Apple, A, B,
# C. D and so on and so forth.

products.sort()
products

# But what we want to do is that we want to uh sort this data
# based upon the prices that we have.
```

```python
products.sort(key=lambda element : element[1])

# So this key will get a lambda and 
# sort based upon is element one, which means
# the prices of each of the product that we have,
# because elements zero is the name of the product and element
# one is the price of each product.

# And then if I check what we have after this, if I run this,
# you can see that I have sorted all of these values based
# upon the prices that we have.

products
```

#### Sorting a list of dictionaries 
- Let's say we have a list of 7 person with their information (name, age, Income) stored in a dictionary.

```python
persons = [
    {"Name" : "Jake" , "Age" : 23 , "Income"  : 450 },
    {"Name" : "Maria" , "Age"  : 32 , "Income"  : 2550 },
    {"Name" : "Will" , "Age"  : 43 , "Income"  : 4500 },
    {"Name" : "Layla" , "Age"  : 12 , "Income"  : 0 },
    {"Name" : "Mina" , "Age"  : 35 , "Income"  : 2100 },
    {"Name" : "Lola" , "Age"  : 28 , "Income"  : 450 },
    {"Name" : "Thelma" , "Age"  : 23 , "Income"  : 450 },
]
```

- sort the list based on each person's age 

```python
persons.sort(key = lambda person : person["Age"])
persons

# And as you can see all of these persons, the list that
# we have is sorted based upon the age of each person.
```

### Task 5: Data Manipulation and Exploration using Lambda Expressions
- we have dataframe containing of information related to 11 Employees who are working in a technology company

```python
import pandas as pd 
dataframe = pd.DataFrame({
    'Name':['Jeremy','Frank','Janet','Ryan','Mary',"John","Alex","Maiki","Alice","Layla","Hanna"],
    'Hourly Wage' : [6,7,22,10,32,10,29,10,12,10,30],
    'Type of contract' : ["Part-Time","Part-Time","Full-Time","Part-Time",
                          "Full-Time","Part-Time","Full-Time","Part-Time",
                          "Part-Time","Part-Time","Full-Time"]
})
dataframe
```

- let's add a column "Daily Working Hours", containing of number of hours each employee works in a day, based on their contract type. 
#### Assume that  : 
    - part-time job working hours : 20 hours per week (4 hours per day)
    - full-time job working hours : 40 hours per week (8 hours per day)

```python
# So what I can do is that there is a method called apply
# that applies a function into each of the elements that I have.

dataframe["Type of contract"].apply(lambda e : 4 if e == "Part-Time" else 8)
```

```python
# Let's create the specific column for it.
dataframe["Daily Working Hours"] = dataframe["Type of contract"].apply(lambda e : 4 if e == "Part-Time" else 8)
```

```python
dataframe
```

- Let's add a column "Annual Income", calculate Annual income of each employee

```python
dataframe.apply(lambda row : row["Daily Working Hours"] * row["Hourly Wage"] * 365, axis=1)
# there is another argument for the apply function.
# So if I hold shift and press tab there is an argument called
# axis.
# So this axis zero for index and one for column default is
# zero.
# So it describes which axis along the function is applied.
# So you have to choose that if you want to apply it
# into the columns or to the rows.
# So what we're gonna do is that we actually want to apply
# this function to each of the rows that we have.
# So we have to put axis equal to one.

```

```python
# create a new column in data frame called and annual income.
dataframe["Annual Income"] = dataframe.apply(lambda row : row["Daily Working Hours"] * row["Hourly Wage"] * 365, axis=1)
```

```python
dataframe
```
